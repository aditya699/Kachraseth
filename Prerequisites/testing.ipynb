# Images in Python

```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def gen_image(path : str):
    img=Image.open(path)
    plt.imshow(img)
    plt.show()
    return img


gen_image("C:/Users/aditya/Desktop/Images/WhatsApp Image 2021-06-29 at 19.58.59.jpeg")
```

```python
# Get image size
img=gen_image("C:/Users/aditya/Desktop/Images/WhatsApp Image 2021-06-29 at 19.58.59.jpeg")
width, height = img.size
print(f"Image size: {width} x {height}")
x=200
y=400
# Get pixel values
pixel_value = img.getpixel((x, y))  # Replace x and y with pixel coordinates
print(f"Pixel value at ({x}, {y}): {pixel_value}")
```
